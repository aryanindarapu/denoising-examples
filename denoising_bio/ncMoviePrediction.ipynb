```python
from __future__ import print_function, unicode_literals, absolute_import, division
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
import glob
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from tifffile import imread
from csbdeep.utils import Path, download_and_extract_zip_file, plot_some
from csbdeep.io import save_tiff_imagej_compatible
from csbdeep.models import CARE
```

```python
class ReadImg:
    def __init__(self, img_name):
        self.img = io.imread(img_name)
        self.__name = img_name

    def __str__(self):
        return self.__name
```

```python
for img in glob.glob("input/*.tif"):
    file = img
    tiffArray = io.imread(file, as_gray=False, plugin=None)

    frameNum = tiffArray.shape[0]
    fHeight = tiffArray.shape[1]
    fWidth = tiffArray.shape[2]

    imArray = np.zeros(shape=(fHeight,fWidth))
    conArray = np.zeros(tiffArray.shape, dtype='uint16')
    
    
    for i1 in range(frameNum):
        imArray = tiffArray[i1]
        io.imsave(('inputFrames/%s.tif'%(i1,)),imArray)
    
    x = imread('inputFrames/1.tif')
    axes = 'YX'
    plt.imshow(x,zorder=1,cmap='gray')
    
    model = CARE(config=None, name='epoch_model', basedir='models')
    
    for i2 in range(frameNum):
        inputImage = imread('inputFrames/%s.tif'%(i2,) )
        restored = model.predict(inputImage, axes)
        Path('outputFrames').mkdir(exist_ok=True)
        save_tiff_imagej_compatible('outputFrames/%s.tif'%(i2,), restored, axes)
        
    for i3 in range(frameNum):
        r_frameNum = 'outputFrames/%s.tif'%(i3,)
        r_frame = io.imread(r_frameNum, as_gray=False, plugin=None)
        conArray[i3] = r_frame


    io.imsave('output/dn_epoch_' + str(file)[6:], conArray)
```
